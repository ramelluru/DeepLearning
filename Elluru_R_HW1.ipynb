```python
# Sigmoid Method
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
## We have created a method inorder to take inputs as Xtrain,Y train,X validation,Y validation ,Learning rate and Iterations
## and returns accuracy of training and validation data,training cost for training and validation data using logistic regression method and sigmoid function.
def model(X, Y,X_val,Y_val, learning_rate, iterations):
    
    #m = X_train.shape[1]
    n = xtrain_1t.shape[0]
    n1 = xval_1t.shape[0]
    W = np.zeros((n,1))
    B = 0
    W1 = np.zeros((n1,1))
    B1 = 0
    cost_list = []
    cost_list1 = []
    acc=[]
    acc1=[]
    for i in range(iterations):
        
        Z = np.dot(W.T, X) + B
        A = sigmoid(Z)
        Z1 = np.dot(W1.T, X_val) + B1
        A1 = sigmoid(Z1)
        cost = -(1/700)*np.sum( Y*np.log(A) + (1-Y)*np.log(1-A)) 
        cost1 = -(1/700)*np.sum( Y_val*np.log(A1) + (1-Y_val)*np.log(1-A1))
        dW = (1/700)*np.dot(A-Y, X.T)
        dB = (1/700)*np.sum(A - Y)
        
        dW1 = (1/700)*np.dot(A1-Y_val, X_val.T)
        dB1 = (1/700)*np.sum(A1 - Y_val)
        
        W = W - learning_rate*dW.T
        B = B - learning_rate*dB
        
        W1 = W1 - learning_rate*dW1.T
        B1 = B1 - learning_rate*dB1
        cost_list.append(cost)
        cost_list1.append(cost1)
        
        A = A > 0.5
    
        A = np.array(A, dtype = 'int64')
    
        ac = (1 - np.sum(np.absolute(A - Y))/Y.shape[1])*100
        
        acc.append(ac)
        
        A1 = A1 > 0.5
    
        A1 = np.array(A1, dtype = 'int64')
    
        ac1 = (1 - np.sum(np.absolute(A1 - Y_val))/Y_val.shape[1])*100
        
        acc1.append(ac1)
        
    return W, B, cost_list,acc,W1,B1,cost_list1,acc1
```

```python
def bmodel(BX, BY,BX_val,BY_val, learning_rate, iterations):

    n = bxtrain_1t.shape[0]
    n1 = bxval_1t.shape[0]
    BW = np.zeros((n,1))
    BB = 0
    BW1 = np.zeros((n1,1))
    BB1 = 0
    Bcost_list = []
    Bcost_list1 = []
    Bacc=[]
    Bacc1=[]
    for i in range(iterations):
        
        BZ = np.dot(BW.T, BX) + BB
        A = sigmoid(BZ)
        BZ1 = np.dot(BW1.T, BX_val) + BB1
        A1 = sigmoid(BZ1)
        cost = -(1/700)*np.sum( BY*np.log(A) + (1-BY)*np.log(1-A))
   
        cost1 = -(1/700)*np.sum( BY_val*np.log(A1) + (1-BY_val)*np.log(1-A1))
        dW = (1/700)*np.dot(A-BY, BX.T)
        dB = (1/700)*np.sum(A - BY)
        
        dW1 = (1/700)*np.dot(A1-BY_val, BX_val.T)
        dB1 = (1/700)*np.sum(A1 - BY_val)
        
        BW = BW - learning_rate*dW.T
        BB = BB - learning_rate*dB
        
        BW1 = BW1 - learning_rate*dW1.T
        BB1 = BB1 - learning_rate*dB1

        Bcost_list.append(cost)
        Bcost_list1.append(cost1)
        
        A = A > 0.5
    
        A = np.array(A, dtype = 'int64')
    
        ac = (1 - np.sum(np.absolute(A - BY))/BY.shape[1])*100
        
        acc.append(ac)
        
        A1 = A1 > 0.5
    
        A1 = np.array(A1, dtype = 'int64')
    
        ac1 = (1 - np.sum(np.absolute(A1 - BY_val))/BY_val.shape[1])*100
        
        acc1.append(ac1)
        
    return BW, BB, Bcost_list,acc,BW1,BB1,Bcost_list1,acc1
```
