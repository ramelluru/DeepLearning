```python
PCA Compression Method: 
I  have  used  the  MNIST  dataset  for  the  PCA  compression  method.  I  have  loaded the data set with 
the  size  28X28  directly  using  the  keras  library  with  google  collab  and  which  will  directly get  the 
training data, testing data, train labels and test labels. Firstly, we shall preprocess all the images by 
dividing  with  255  and  converting  to  float32  and  making  all  the  pixel  values  in  between  0&1.  We 
shall apply the PCA algorithm the unsupervised learning algorithm which takes the correlated  input 
and  provide  the  uncorrelated  outputs.  We  shall  apply  PCA  with the number of components as 784 
which  is  the  raw  image  dimensions  and  get  the  PCA  compressed  data  and  fed  into  our  model  and 
calculate the test accuracy which is our comparison factor in this experiment. We shall continue the 
experiment  with the  number  of  components  as  the  500,100,50,10  and  1  which  shall  compress  the 
images  with  the  known  proportions  as  specified  and  continue  with  the  experiment  as  mentioned 
above proceed with the deliverable i.e., test accuracy. We are also tracking the graphical 
representation of train and validation accuracies and losses. 
```

```python
## Importing libraries
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import keras
import seaborn as sns
from sklearn.metrics import confusion_matrix,accuracy_score,ConfusionMatrixDisplay
from keras.datasets import mnist
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
from sklearn.decomposition import PCA
```

```python
## Fetching the MNIST data set and assiging test and train data
(xtrain, ytrain), (xtest,ytest) = keras.datasets.mnist.load_data()
## Reshaping the datasets for further operations
size = xtrain.shape[1] * xtrain.shape[2]
xtrain = xtrain.reshape(xtrain.shape[0], size)
xtest = xtest.reshape(xtest.shape[0], size)
## Normalizing datasets
xtrain = xtrain.astype('float32')/255.0
xtest=xtest.astype('float32')/255.0
## Plotting the images
x1 = xtrain[1,:].reshape([28,28])
plt.imshow(x1, cmap='gray_r')
plt.show()
x2 = xtest[1,:].reshape([28,28])
plt.imshow(x2, cmap='gray_r')
plt.show()
```

```python
## Setting the components that we are evaluvating PCA and applying the CNN and getting the parameters
Components=[784,500,100,50,10,1]
for i in range(len(Components)):
    print("PCA with number of components:",Components[i])
    ## Applying PCA
    pca = PCA(n_components=Components[i])
    compressed = pca.fit_transform(xtrain)
    recovered = pca.inverse_transform(compressed)
    recoveredx1 = recovered[1,:].reshape([28,28])
    plt.imshow(recoveredx1, cmap='gray_r')
    plt.show()
    pca = PCA(n_components=Components[i])
    compressedtest = pca.fit_transform(xtest)
    recoveredtest = pca.inverse_transform(compressedtest)
    recoveredx2 = recoveredtest[1,:].reshape([28,28])
    plt.imshow(recoveredx2, cmap='gray_r')
    plt.show()
    ## Creating Model
    imageinput = Input(shape=(784,))
    x = Dense(32,activation='relu')(imageinput)
    y = Dense(784, activation='sigmoid')(x)

    autoencoder = Model(imageinput, y)
    ## Compiling the model
    autoencoder.compile(optimizer='adam',loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
    ## Fitting the model
    history=autoencoder.fit(recovered, ytrain,epochs=10,batch_size=128,validation_data=(recoveredtest, ytest))
    y_pred1 = autoencoder.predict(recoveredtest)
    y_pred1 = np.argmax(y_pred1, axis=1)

    ## Calculating Test Accuracy
    print("Test Accuracy obtained:",accuracy_score(ytest, y_pred1))

    ## Plotting the Train and Validation losses and Accuracies
    plt.plot(history.history['loss'], color='r',label='train loss')
    plt.plot(history.history['val_loss'], color='b', label ='val loss')
    plt.title("Loss")
    plt.legend()
    plt.show()
    plt.plot(history.history['accuracy'], color='r', label="train accuracy")
    plt.plot(history.history['val_accuracy'], color='b', label="val accuracy")
    plt.title("Accuracy")
    plt.legend()
    plt.show()
```
